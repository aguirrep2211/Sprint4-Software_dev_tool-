```python
# Crear directorio para guardar la notebook
/home/aguirre_p/tripleten/sprint4/Sprint4-Software_dev_tool-/notebooksnotebook_dir = ""
os.makedirs(notebook_dir, exist_ok=True)
```

```python
# Save the notebook with visualizations
notebook_code = f"""
import pandas as pd
import plotly.express as px

# Cargar datos
df = pd.read_csv("../vehicles_us.csv")

# Histograma del precio
fig_price_hist = px.histogram(df, x="price", nbins=100, title="DistribuciÃ³n de Precios")
fig_price_hist.show()

# Histograma del kilometraje
fig_odometer_hist = px.histogram(df, x="odometer", nbins=100, title="DistribuciÃ³n del Kilometraje")
fig_odometer_hist.show()

# Scatterplot: precio vs aÃ±o del modelo
fig_price_model_year = px.scatter(df, x="model_year", y="price", title="Precio vs AÃ±o del Modelo", opacity=0.5)
fig_price_model_year.show()

# Scatterplot: precio vs kilometraje
fig_price_odometer = px.scatter(df, x="odometer", y="price", title="Precio vs Kilometraje", opacity=0.5)
fig_price_odometer.show()
"""

notebook_path = f"{notebook_dir}/EDA.ipynb"
```

```python
# Crear el notebook
nb = nbf.v4.new_notebook()
nb['cells'] = [nbf.v4.new_code_cell(notebook_code)]
```

```python
# Guardar el archivo
notebook_path = os.path.join(notebook_dir, "EDA.ipynb")
with open(notebook_path, 'w') as f:
    nbf.write(nb, f)
```

```python
import streamlit as st

# Encabezado principal
st.header("ExploraciÃ³n de Datos de VehÃ­culos en EE.UU.")

# Histograma del precio
st.subheader("DistribuciÃ³n de Precios")
st.plotly_chart(px.histogram(vehicles_df, x="price", nbins=100, title="DistribuciÃ³n de Precios"))

# Scatterplot: precio vs kilometraje
st.subheader("Precio vs Kilometraje")
st.plotly_chart(px.scatter(vehicles_df, x="odometer", y="price", title="Precio vs Kilometraje", opacity=0.5))

# Checkbox para mostrar un histograma alternativo por aÃ±o del modelo
if st.checkbox("Mostrar histograma por aÃ±o del modelo"):
    st.subheader("DistribuciÃ³n de AÃ±o del Modelo")
    st.plotly_chart(px.histogram(vehicles_df, x="model_year", nbins=40, title="DistribuciÃ³n por AÃ±o del Modelo"))
```
